```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# generating sample data
np.random.seed(0)
num_bedrooms = np.random.randint(1,6,100)
# house prices (per thousand)
house_price = 50 * num_bedrooms + np.random.normal(0,10,100)
```

```python
# Splitting the data into train and testing sets
x_train,x_test,y_train,y_test = train_test_split(num_bedrooms.reshape(-1,1),house_price,test_size=0.2)
```

```python
# create linear regression model
r_model = LinearRegression()
```

```python
# fit the model on the training data
r_model.fit(x_train, y_train)
```

```python
# Making prediction on the test data
y_pred = r_model.predict(x_test)
```

```python
# Calculating mean squared Error
mse = mean_squared_error(y_test,y_pred)
```

```python
# Printing the model parameters
print("Intercept", r_model.intercept_)
print("Coefficient", r_model.coef_[0])
print("Mean Squared Error", mse)
```

```python
# Plotting the data and regression line
plt.scatter(x_test,y_test, color="blue")
plt.plot(x_test,y_pred, color='red', linewidth=2)
plt.xlabel("Number of Bedrooms")
plt.ylabel("House Price(in thousands)")
plt.title("Linear Regression Model")
plt.show()
```
